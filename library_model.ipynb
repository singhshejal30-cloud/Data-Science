```python
import pandas as pd
```

```python
data = pd.read_csv("library_data_100.csv")
data
```

```python
print(data.groupby("FrequentUser").mean())
```

```python
X=data[["StudentAge","BooksIssued","LateReturns","MembershipYears"]]
y=data["FrequentUser"]
```

```python
#labelEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
le=LabelEncoder()
y_enc=le.fit_transform(y)
```

```python
#trainng = 80% data & testing =20% data
X_train,X_test,y_train,y_test=train_test_split(X,y_enc,test_size=0.2,random_state=42)
```

```python
#train Model
from sklearn.tree import DecisionTreeClassifier
```

```python
#80% data
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#20% data for prediction or testing
pred=model.predict(X_test)
```

```python
#check accuracy of the data
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,pred)
print("Accuracy of model:",accuracy * 100,"%")
```

```python
new_student = [[22, 12, 1, 3]]
prediction = model.predict(new_student)

if prediction[0] == 1:
    print("Prediction: Frequent Library User")
else:
    print("Prediction: Not a Frequent Library User")
```

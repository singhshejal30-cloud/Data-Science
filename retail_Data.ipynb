```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import data from csv and store in variable df
df=pd.read_csv("retail_data.csv")
df               
               
```

```python
# information of database
df.info()
```

```python
# statics summery
df.describe()
```

```python
# check null values
df.isnull().sum()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.groupby("Age")["Customer ID"].count()
```

```python
# convert date column to datetime standerd
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
# create month column
df["Month"]=df["Date"].dt.to_period("M")
df
```

```python
#create graph
monthly_sales=df.groupby("Month")["Total Amount"].sum()

plt.figure()
monthly_sales.plot(kind="bar",color="brown")
plt.title("monthly sales")
plt.xlabel("Month")
plt.ylabel("Total sales")
plt.show()  #close the code
```

```python
cat_sales=df.groupby("Product Category")["Total Amount"].sum()
colors=["pink","green","orange"]
plt.figure(figsize=(5,2))  #create a new graph
cat_sales.plot(kind="bar",color=colors)
plt.title("Product Category sales of retail data")
plt.xlabel("Product Category")
plt.ylabel("Total sales")
plt.show()  #close the code
```

```python
plt.figure()
plt.scatter(df["Quantity"],df["Total Amount"],color="green")
plt.title("Quantity vs Total Amount")
plt.xlabel("Quantity")
plt.ylabel("Total Amount")
plt.show()
```

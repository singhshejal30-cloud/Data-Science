```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Chocolate_sales.csv")
```

```python
df
```

```python
df.info()
```

```python
# convert date column to datetime standerd
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
df["Month"]=df["Date"].dt.to_period("M")
df
```

```python
#create graph
monthly_sales=df.groupby("Sales Person")["Amount"].sum()

plt.figure()
monthly_sales.plot(kind="bar",color="brown")
plt.title("Sales bt Sales Person")
plt.xlabel("Sales Person")
plt.ylabel("Total sales Amount")
plt.show()  #close the code
```

```python
cat_sales=df.groupby("Country")["Amount"].sum()
colors=["pink","black","brown"]
plt.figure(figsize=(5,5))  #create a new graph
cat_sales.plot(kind="bar",color=colors)
plt.title("Sales by country")
plt.xlabel("Total Sales Amount")
plt.ylabel("Country")
plt.show()  #close the code
```

```python
#create graph
monthly_sales=df.groupby("Product")["Amount"].sum()

plt.figure()
monthly_sales.plot(kind="bar",color="black")
plt.title(" Sales by Person")
plt.xlabel("Product")
plt.ylabel("Total sales Amount")
plt.show()  #close the code
```

```python
plt.figure()
plt.scatter(df["Product"],df["Amount"],color="yellow")
plt.title("Product vs Total Amount")
plt.xlabel("Product")
plt.ylabel("Amount")
plt.show()
```

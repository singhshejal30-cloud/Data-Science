```python
#SVD- Singular Value Decomposition

import numpy as np
```

```python
marks= np.array([[80,90,np.nan],
                 [60,np.nan,70],
                 [90,85,88],
                 [np.nan,55,60],
                 [75,np.nan,np.nan]])
```

```python
marks
```

```python
upd_marks=np.nanmean(marks,axis=1)
```

```python
upd_marks
```

```python
#Replace NaN with student-wise mean
filled_marks=np.where(np.isnan(marks),upd_marks[:,np.newaxis],marks)
```

```python
filled_marks
```

```python
#Perform SVD
U,S,VT=np.linalg.svd(filled_marks,full_matrices=False)
```

```python
#Reconstruct approximate marks matrix
k=2 #it will keep the top 2 most imp marks pattern

U_k = U[:, :k] #Most imp student pattern
S_k = np.diag(S[:k]) #how much is the strength of the pattern
VT_k = VT[:k, :] #Most imp subject in the dataset

approx_marks = U_k @ S_k @ VT_k #contruct the matrix with most imp features
```

```python
approx_marks
```
